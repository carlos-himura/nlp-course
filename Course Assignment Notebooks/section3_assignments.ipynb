```python
def preprocess_text(text_series):
    text_series = text_series.str.lower() # Convert to lowercase
    text_series = text_series.str.replace("\xa0", " ", regex=False).str.strip() # Replace non-breaking spaces with regular spaces and strip whitespace
    text_series = text_series.str.replace(r'[^\w\s]', '', regex=True) # Remove punctuation
    return text_series
```

```python
books_df["Description_Clean"] = preprocess_text(books_df["Description"])
```

## 2. Text Preprocessing with spaCy

In addition to the lowercasing and special character removal from the previous assignment, within the cleaned Description column:
* Tokenize the text
* Lemmatize the text
* Remove stop words

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
def lower_replace(series):
    output = series.str.lower()
    output = output.str.replace(r'\[.*?\]', '', regex=True)
    output = output.str.replace(r'[^\w\s]', '', regex=True)
    return output

def token_lemma_nonstop(text):
    doc = nlp(text)
    output = [token.lemma_ for token in doc if not token.is_stop] # Remove stop words
    output = ' '.join(output) # Join tokens back into a single string
    return output 

def filter_pos(text, pos_list=['NOUN', 'PROPN']):
    doc = nlp(text)
    output = [(token.text) for token in doc if token.pos_ in pos_list] # Filter by part of speech
    output = ' '.join(output)
    return output
```

```python
# create an nlp pipeline
def nlp_pipeline(series):
    output = lower_replace(series)
    output = output.apply(token_lemma_nonstop)
    return output
```

```python
books_df.head(5)
```

```python
text_clean = nlp_pipeline(books_df['Description_Clean'])
text_clean.head(10)
```

```python
books_df.Description_Clean = text_clean
books_df.head(5)
```

```python
pd.to_pickle(text_clean, 'text_clean.pkl') # Save the cleaned text to a pickle file
```

## 3. Count Vectorizer

1. Vectorize the cleaned and normalized text using Count Vectorizer with the default parameters
2. Modify the Count Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
3. Use the updated Count Vectorizer to identify the:
* Top 10 most common terms
* Top 10 least common terms that appear in at least 10% of the documents
4. Create a horizontal bar chart of the top 10 most common terms

```python
# if you don't want to run all your cleaning code again, just load in the pickle file
series = pd.read_pickle('text_clean.pkl')
series
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer() # initialize the CountVectorizer
dtm = cv.fit_transform(series) # create document-term matrix
dtm_df = pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out()) # convert to DataFrame
dtm_df
```

```python
# count vectorizer code with parameter tweaks
cv2 = CountVectorizer(stop_words='english', min_df=.1) # remove English stop words and terms that appear in less than 10% of documents
dtm2 = cv2.fit_transform(series) # create document-term matrix
dtm_df2 = pd.DataFrame(dtm2.toarray(), columns=cv2.get_feature_names_out()) # convert to DataFrame
dtm_df2
```

```python
# return the most common terms
term_freq = dtm_df2.sum()
term_freq
```

```python
# get the top 10 terms
top_10_terms = term_freq.sort_values(ascending=False).head(10)
top_10_terms
```

```python
# get the least 10 terms
least_10_terms = term_freq.sort_values(ascending=True).head(10)
least_10_terms
```

```python
top_10_terms.sort_values().plot(kind='barh');
```

## 4. TF-IDF Vectorizer

1. Vectorize the cleaned and normalized text using TF-IDF Vectorizer with the default parameters
2. Modify the TF-IDF Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
* Set a maximum document frequency of 50%
3. Using the updated TF-IDF Vectorizer, create a  horizontal bar chart of the top 10 most highly weighted terms
4. Compare the Count Vectorizer bar chart from the previous assignment with the TF-IDF Vectorizer bar chart and note the differences in the top term lists

```python
# basic tfidf vectorizer code
# How important is a word to THIS document, compared to the ENTIRE corpus?
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer() 
tfidf = tv.fit_transform(series)
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tv.get_feature_names_out())
tfidf_df
```

```python
# tfidf vectorizer with some parameter tweaks
tv2 = TfidfVectorizer(stop_words='english', min_df=.1, max_df=.5) # remove English stop words, terms that appear in less than 10% of documents, and terms that appear in more than 50% of documents
tfidf2 = tv2.fit_transform(series)
tfidf_df2 = pd.DataFrame(tfidf2.toarray(), columns=tv2.get_feature_names_out())
tfidf_df2
```

```python
# return the most common terms
term_freq_tf_idf = tfidf_df2.sum()
term_freq_tf_idf
```

```python
# get the top 10 terms
top_10_terms_tf_idf = term_freq_tf_idf.sort_values(ascending=False).head(10)
top_10_terms_tf_idf
```

```python
top_10_terms_tf_idf.sort_values().plot(kind='barh');
```

```python
# Comparison bwteen Count Vectorizer and TF-IDF
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# TF-IDF top 10
top_10_terms_tf_idf.sort_values().plot(
    kind='barh',
    ax=axes[0],
    title="Top 10 Terms (TF-IDF)"
)

# CountVectorizer top 10
top_10_terms.sort_values().plot(
    kind='barh',
    ax=axes[1],
    title="Top 10 Terms (Raw Frequency)"
)

plt.tight_layout()
plt.show()
```

```python
# differences between the two methods
# CountVectorizer highlights the most frequently occurring words in the corpus.
# TF-IDF highlights words that are important to specific documents but not common across all documents.
```
